# Linear Regression

Let's fabricate some data that shows a roughly linear relationship between page speed and amount purchased:

```python
%matplotlib inline
import numpy as np
from pylab import *

pageSpds = np.random.normal(3.0, 1.0, 1000)
purchaseAmt = 100 - (pageSpds + np.random.normal(0, 0.1, 1000)) * 3

scatter(pageSpds, purchaseAmt)
```

As we only have two features, we can keep it simple and just use scipy.state.linregress:

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpds, purchaseAmt)
```

Not surprisngly, our R-squared value shows a really good fit:

```python
r_value ** 2
```

Let's use the slope and intercept we got from the regression to plot predicted values vs. observed:

```python
import matplotlib.pyplot as plt

def predict(x):
    return slope * x + intercept

fitLine = predict(pageSpds)

plt.scatter(pageSpds, purchaseAmt)
plt.plot(pageSpds, fitLine, c='r')
plt.show()
```
